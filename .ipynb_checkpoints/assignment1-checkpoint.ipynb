**Question 2**: Use a plotting utility (e.g. `matplotlib.pyplot`) to plot the arrival times of elevators on the training data
(Monday 9/16 - Tuesday 9/17). The x-axis should be timestamp and the y-axis should be the elevator ID. I suggest using a scatter plot.
ChatGPT may be useful. Display the plot. List any observations
(e.g. cyclical patterns, some elevators arriving more frequently than others)

```python
# Code for question 1:
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
from datetime import timedelta
import numpy as np


data = data = pd.read_csv('data.csv')

# Convert the timestamp to a datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
    
# Filter the data to include only records on or before 9/17/2024
training_data = data[data['Timestamp'] <= '2024-09-17 23:59:59']


def make_plot(data):
    '''
    You will need to read the training data CSV file and do some processing first.
    '''

    # Plotting the data using a scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(data['Timestamp'], data['Elevator ID'], alpha=0.6, color='b')

    # Formatting the x-axis for better readability
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d %H:%M'))
    plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=1))
    plt.gcf().autofmt_xdate()

    # Adding labels and title
    plt.title('Elevator Arrival Times (Monday 9/16 - Tuesday 9/17)')
    plt.xlabel('Timestamp')
    plt.ylabel('Elevator ID')

    # Display the plot
    plt.show()

make_plot(training_data)
```

```python
# Ensure data is sorted by Timestamp
filtered_training_data = filtered_training_data.sort_values(by='Timestamp')

# Calculate the difference between consecutive timestamps
filtered_training_data['Time Difference'] = filtered_training_data['Timestamp'].diff()

# Identify gaps larger than 5 minutes
large_gaps = filtered_training_data[filtered_training_data['Time Difference'] > timedelta(minutes=5)]

# For each large gap, create a DataFrame that shows the "end of one period" and "start of the next"
filtered_periods = pd.DataFrame({
    'Gap Start': large_gaps['Timestamp'] - large_gaps['Time Difference'],
    'Gap End': large_gaps['Timestamp'],
    'Gap Duration': large_gaps['Time Difference']
})

# Display the filtered out periods
print(filtered_periods)

# Set a threshold of 5 minutes (300 seconds) and filter out gaps greater than 5 minutes
filtered_training_data['Valid Difference'] = filtered_training_data['Time Difference'].apply(lambda x: x if x <= timedelta(minutes=5) else timedelta(0))

# Calculate total valid usage time in seconds
total_usage_time = filtered_training_data['Valid Difference'].sum().total_seconds()

# Display the total usage time in seconds
total_usage_time
```

```python
positions = np.array([(1, 3), (3, 3), (5, 3), (5, 0), (3, 0), (1, 0)])
actual_ideal_position = np.average(positions, axis=0, weights=arrival_counts)
actual_ideal_position
```

**Question 6**: On the training data, calculate the average distance traveled to the next elevator arrival if I wait at the location specified in question 1.
Then, calculate the average distance traveled if I wait at the location specified in question 5.

*Your answer to question 5 should result in a shorter average distance traveled.*

You are free to write your own code for these calculations, but we recommend implementing a function called `get_average_walk_distance(data, coordinate)`, since you will need to reuse this function.



```python
# Your code here

def get_average_walk_distance(data, centroid):
    '''
    data is a csv file containing the data.
    coordinate is the coordinate to wait at.
    '''
    ### TODO
    # Mapping of elevator positions
    elevator_positions = {
        1: (1, 3),
        2: (3, 3),
        3: (5, 3),
        4: (5, 0),
        5: (3, 0),
        6: (1, 0)
    }
    
    data = data.copy()
    
    # Ensure centroid is an array
    centroid = np.array(centroid)

    # Map elevator IDs to coordinates
    data['Coordinates'] = data['Elevator ID'].map(elevator_positions)

    # Calculate the distance to the input centroid for each elevator arrival
    data['Distance to Centroid'] = data['Coordinates'].apply(lambda coords: np.linalg.norm(np.array(coords) - centroid))

    # Calculate the average walking distance
    average_walk_distance = data['Distance to Centroid'].mean()
    return average_walk_distance


print(np.round(get_average_walk_distance(filtered_training_data, (3,1.5)),5))
print(np.round(get_average_walk_distance(filtered_training_data, actual_ideal_position),5))
```
